```
import pandas as pd
import numpy as np
import sklearn
```

```
#load dataset with pandas
msrm_df=pd.read_csv('/content/drive/MyDrive/Ai-datasets/mushroom edibility classification dataset.csv')
msrm_df.head()
```

```
msrm_df.shape
```

```
#handling missing values
msrm_df.isnull().sum()
```

```
from sklearn.impute import SimpleImputer
impute = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
impute.fit(msrm_df[['cap-shape']])
msrm_df['cap-shape']=impute.transform(msrm_df[['cap-shape']])
```

```
impute.fit(msrm_df[['cap-color']])
msrm_df['cap-color']=impute.transform(msrm_df[['cap-color']])
```

```
msrm_df.isnull().sum()
```

```
# Encoding categorical features
msrm_df.info()
```

```
msrm_df['bruises'].unique()
```

```
msrm_df['bruises']=msrm_df['bruises'].map({'bruises is there':1, 'no bruises':0})
```

```
msrm_df.info()
```

```
msrm_df['ring-number'].unique()
```

```
msrm_df.drop(['ring-number'], axis=1, inplace=True)
msrm_df.info()
```

```
msrm_df['veil-color'].unique()
```

```
msrm_df.drop(['veil-color'], axis=1, inplace=True)
msrm_df.info()
```

```
msrm_df['veil-type'].unique()
```

```
msrm_df.drop(['veil-type'], axis=1, inplace=True)
msrm_df.info()
```

```
msrm_df.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```
#logistic regression Start
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```
msrm_df_reg=msrm_df
msrm_df_reg.head()
```

```
msrm_df_reg.shape
```

```
Y=msrm_df_reg.iloc[:,0] #target values
X=msrm_df_reg.iloc[:,1:16] #feature values
```

```
print(Y.head())
print(X.head())
```

```
#Split the data into 80% training and 20% testing
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=10)
```

```
#Training the model
model = LogisticRegression()
model.fit(x_train, y_train) #Training the model
predictions = model.predict(x_test)
print(predictions)# printing predictions
```

```
print(accuracy_score(y_test, predictions))
log_acc=accuracy_score(y_test, predictions)
#logistic regression end
```

```
#decision tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
Y=msrm_df_reg.iloc[:,0] #target values
X=msrm_df_reg.iloc[:,1:16] #feature values
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=10)
clf = DecisionTreeClassifier(criterion='entropy',random_state=1)
clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
score=accuracy_score(y_pred,y_test)
decs_score=score
print(score)
```

```
#checking the decision tree
from sklearn import tree
import matplotlib.pyplot as plt
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (3,3), dpi=300)
tree.plot_tree(clf,
               feature_names = X.columns, 
               class_names=Y.unique(),
               filled = True);
```

```
print(log_acc)
print(decs_score)
```

```
score = [log_acc, decs_score]
range = [0,1]
colors = ['red','blue']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (10,10), dpi=100)
axes.set_yticks(np.linspace(0,1, num=50))
axes.set_xticks([0,1])
axes.set_xticklabels(['Logistic Regression','Decision tree'])
plt.bar(range, score, color=colors, width=0.5)
plt.title('Accuracy Compare')
plt.xlabel('Accuracy Classification')
plt.ylabel('Score')
plt.show()
```
